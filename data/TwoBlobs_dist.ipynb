```python
import numpy as np
from sklearn.datasets import make_blobs

def make_blobs_dist(n_samples, dist=0, seed=0):
    np.random.seed(seed)
    X, l = make_blobs(n_samples, 2, centers=np.array([[0-dist, 0], [0+dist, 0]]), cluster_std=[1, 1])
    return X, l


np.random.seed(0)
random_seeds = np.random.choice(1_000_000, replace=False, size=10)
dists = np.arange(0, 15.1, 0.5)
blobs_datasets = [[make_blobs_dist(500, dist=dist, seed=seed) for seed in random_seeds] for dist in dists]
```

```python
import matplotlib.pyplot as plt


X, l = make_blobs_dist(500, 0, 0)

plt.figure(figsize=(5, 5))
color = plt.cm.tab20(np.linspace(0, 1, len(set(l)) + 1))
for c_id in np.sort(list(set(l))):
    plt.scatter(
        X[:, 0][l == c_id], X[:, 1][l == c_id], s=5, color=color[int(c_id) + 1], label=int(c_id)
    )
legend = plt.legend(
    loc="center left",
    bbox_to_anchor=(1, 0.5),
)
ax = plt.gca()
plt.ylim(-10, 10)
plt.xlim(-10, 10)
ax.set_aspect("equal", adjustable="datalim")
```
