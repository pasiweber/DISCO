```python
from sklearn.datasets import make_moons
from data._utils import add_noise

def make_moons_random_assignment(n_samples, density=1, seed=0):
    np.random.seed(seed)
    X, l = make_moons(n_samples=(n_samples, n_samples*density), random_state=seed, noise=0.0)
    # X, l = add_noise(X, l, 20, 0.1)
    class0 = np.where(l == 0)[0]
    class1 = np.where(l == 1)[0]
    # choice_class0 = np.random.choice(len(class0), replace=False, size=n_samples // 100 * n_swap)
    # choice_class1 = np.random.choice(len(class1), replace=False, size=n_samples // 100 * n_swap)
    # l[class0[choice_class0]] = 1
    # l[class1[choice_class1]] = 0
    return X, l

### twoMoonsDatasets[n_dims][run]
densities = list(range(1, 11, 1))
np.random.seed(0)
random_seeds = np.random.choice(1_000_000, replace=False, size=10)
twoMoonsDatasets = [
    [make_moons_random_assignment(100, density=density, seed=seed) for seed in random_seeds]
    for density in densities
]
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(2*5, 2*5))
G = gridspec.GridSpec(5, 5)

for density in range(0, len(densities)):
    ax = plt.subplot(G[density // 5, density % 5])
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title(f"{densities[density]}")
    X, l = twoMoonsDatasets[density][0]
    plt.scatter(X[::2, 0], X[::2, 1], s=1, c=l[::2])

plt.savefig("imgs/TwoMoons_density_data.png")
```

```python
import cache_magic

import time
from collections import defaultdict
from data._metrics import METRICS
from data._utils import insert_dict


def calc():
    eval_results = defaultdict(list)

    for density in range(len(twoMoonsDatasets)):
        for run in range(len(twoMoonsDatasets[0])):
            X, l = twoMoonsDatasets[density][run]
            for name, metric_fn in METRICS.items():
                start_time = time.time()
                start_process_time = time.process_time()
                value = metric_fn(X, l)
                end_process_time = time.process_time()
                end_time = time.time()
                insert_dict(eval_results, {
                    "measure": name, 
                    "density": densities[density],
                    "run": run,
                    "value": value,
                    "time": end_time - start_time,
                    "process_time": end_process_time - start_process_time,
                })
    return eval_results

%cache eval_results_density = calc()
```

```python
import pandas as pd

SELECTED_MEASURES = [
    "DISCO",
    # "DC_DUNN",
    ### Competitors
    "DBCV",
    "DCSI",
    # "S_DBW",
    "CDBW",
    "CVDD",
    # "CVNN",
    "DSI",
    ### Gauss
    "SILHOUETTE",
    "DUNN",
    # "DB",
    # "CH",
]

df = pd.DataFrame(data=eval_results_density)
df = df[df.measure.isin(SELECTED_MEASURES)]
df["measure"] = pd.Categorical(df["measure"], SELECTED_MEASURES)
df = df.sort_values("measure")
# df.groupby("measure").sum()
```
