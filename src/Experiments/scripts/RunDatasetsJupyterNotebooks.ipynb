```python
import glob
from PIL import Image, ImageOps

def crop_png(path):
    im = Image.open(path)
    im = im.convert("RGB")
    invert_im = ImageOps.invert(im)
    im2 = im.crop(invert_im.getbbox())
    im2.save(path)

for path in glob.glob("imgs/**/*.png", recursive=True):
    crop_png(path)
```
