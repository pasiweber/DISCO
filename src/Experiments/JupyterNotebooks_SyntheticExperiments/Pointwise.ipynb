```python
from datasets.density_datasets import Datasets as DensityDatasets
from clustpy.utils import plot_with_transformation
import matplotlib.pyplot as plt
from src.Evaluation.disco import disco_samples
import matplotlib.colors as colors
import numpy as np
from sklearn.cluster import KMeans


def truncate_colormap(cmap, minval=0.0, maxval=1.0, n=100):
    new_cmap = colors.LinearSegmentedColormap.from_list(
        "trunc({n},{a:.2f},{b:.2f})".format(n=cmap.name, a=minval, b=maxval),
        cmap(np.linspace(minval, maxval, n)),
    )
    return new_cmap
```

```python
from matplotlib.colors import LinearSegmentedColormap
colors = [(0, "darkred"),(0.2, "darkred"), (0.5, "rosybrown"), (0.8, "darkslategrey"),(1, "darkslategrey")]
cmap = LinearSegmentedColormap.from_list("custom_cmap", colors)
cmap
```

```python
X_n, l_n = DensityDatasets.cluto_t7_10k.standardized_data_cached_no_noise
np.random.seed(0)
seeds = np.random.choice(10_000, size=1, replace=False)
np.random.seed(seeds[-1])
shuffle_data_index = np.random.choice(len(X_n), size=len(X_n), replace=False)
X_n = X_n[shuffle_data_index]
l_n = l_n[shuffle_data_index]

kmeans = KMeans(len(set(l_n)))
l_kmeans = kmeans.fit_predict(X_n)

disco_gt_n = disco_samples(X_n,l_n)
disco_kmeans_n = disco_samples(X_n, l_kmeans)
```

```python
plt.scatter(X_n[:, 0], X_n[:, 1], s=3,vmin=-1, vmax=1, c=disco_gt_n, cmap=cmap)
plt.xlim(np.min(X_n[:, 0]-0.2), np.max(X_n[:, 0])+0.2)
plt.ylim(np.min(X_n[:, 1]-0.2), np.max(X_n[:, 1])+0.2)
plt.xticks([])
plt.yticks([])
#plt.savefig("complex9_gt.png", bbox_inches="tight",dpi=300, pad_inches=0)
```

```python
from scipy.spatial import Voronoi, voronoi_plot_2d
# compute Voronoi tesselation
points = kmeans.cluster_centers_
vor = Voronoi(points)
# plot


voronoi_plot_2d(vor, show_points=False, show_vertices=False)
plt.scatter(X_n[:, 0], X_n[:, 1], s=3,vmin=-1, vmax=1, c=disco_kmeans_n, cmap=cmap)
plt.xlim(np.min(X_n[:, 0]-0.2), np.max(X_n[:, 0])+0.2)
plt.ylim(np.min(X_n[:, 1]-0.2), np.max(X_n[:, 1])+0.2)
plt.xticks([])
plt.yticks([])
#plt.savefig("complex9_kmeans.png", bbox_inches="tight",dpi=300, pad_inches=0)
```
