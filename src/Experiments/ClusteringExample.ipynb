```python
import pandas as pd
import numpy as np
from datasets.density_datasets import Datasets as DensityDatasets
from ast import literal_eval

DISCO_ROOT_PATH = "/export/share/pascalw777dm/DISCO/"
from datasets.density_datasets import Datasets as DensityDatasets


def load_clustering(args):
    dataset, clusterer, run = args
    X, l = dataset.standardized_data_cached

    np.random.seed(0)
    seeds = np.random.choice(10_000, size=run + 1, replace=False)
    np.random.seed(seeds[-1])
    shuffle_data_index = np.random.choice(len(X), size=len(X), replace=False)
    X = X[shuffle_data_index]
    l = l[shuffle_data_index]

    df = pd.read_csv(f"{DISCO_ROOT_PATH}/clusterings_/{dataset.id}/{clusterer}_{run}.csv")
    l_ = df["value"][0]
    l_ = np.array(literal_eval(",".join(l_.split()).replace("[,", "[")))

    return X, l, l_

three_datasets_gt = [[load_clustering((DensityDatasets.three_spiral, "KMeans", i))[0:2]] for i in range(10)]
three_datasets_kmeans = [[load_clustering((DensityDatasets.three_spiral, "KMeans", i))[0:3:2]] for i in range(10)]
complex9_datasets_gt = [[load_clustering((DensityDatasets.complex9, "KMeans", i))[0:2]] for i in range(10)]
complex9_datasets_kmeans = [[load_clustering((DensityDatasets.complex9, "KMeans", i))[0:3:2]] for i in range(10)]
```
